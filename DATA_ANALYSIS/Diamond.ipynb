```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Loading the diamong dataset from seaborn
df=sns.load_dataset('diamonds')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.hist(df['price'],bins=30,color='skyblue')
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Price")
plt.show()
```

```python
plt.scatter(x=df['carat'],y=df['price'])
plt.show()
```

```python
avg_across_cuts=df.groupby('cut')['price'].mean()
print(avg_across_cuts)
```

```python
avg_across_cuts.plot(kind='bar')
plt.show()
```

```python
carat_price=df[['carat','price']]
carat_price_corr=carat_price.corr()
print(carat_price_corr)
```

```python
sns.lineplot(x=df['carat'],y=df['price'],data=df)
plt.show()
```

```python
avg_price_based_clarity=df.groupby('clarity')['price'].mean()
print(avg_price_based_clarity)
avg_price_based_clarity.plot(kind='bar')
plt.show()
```

```python
sns.lineplot(x=df['carat'],y=df['cut'],data=df)
plt.show()
```

```python
cut_vs_carat=df.groupby('cut')['carat'].mean()
print(cut_vs_carat)
cut_vs_carat.plot(kind='bar')
plt.show()
```

```python
sns.boxplot(data=df,x='cut',y='carat')
plt.show()
```

```python
cut_color_price=df.groupby(['cut','color'])['price'].mean()
print(cut_color_price)
cut_color_price.plot(kind='bar')
plt.show()
```

```python
pivot_table=pd.pivot_table(df,values='price',index='cut',columns='color',aggfunc='mean')
print(pivot_table)
                           
```

```python
Q1=df['price'].quantile(0.25)
Q2=df['price'].quantile(0.50)
Q3=df['price'].quantile(0.75)
Q4=df['price'].quantile(1)

print(Q1,Q2,Q3,Q4)
```

```python
IQR=Q3-Q2
print(IQR)
```

```python
outliers=(df['price']<Q1-1.5*IQR) |(df['price']<Q3+1.5*IQR)
print(df[outliers])
```

```python
df.info()
```

```python
price_quartile=pd.qcut(df['price'],q=4,labels=['Q1','Q2','Q3','Q4'])
df['price_quartile']=price_quartile

sns.boxplot(data=df,x='price_quartile',y='carat',order=['Q1', 'Q2', 'Q3', 'Q4'])
plt.show()
```

```python
df['price_per_carat']=df['price']/df['carat']

plt.hist(df['price_per_carat'])
plt.show()
```
